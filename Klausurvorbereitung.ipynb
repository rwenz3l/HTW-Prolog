### Wiederholung Inferenzregeln:

- Bilden Logischer Zusammenhänge & Schlussfolgerungen

**modus ponens**:
 - P → Q ist eine Beziehung von 2 Fakten [ z.B.: "Wenn es regnet, wird die Straße nass" ]
 - Es lässt sich Schlussfolgern, dass wenn P eintritt, auch Q eintritt.
 - Wenn also P wahr ist, dann ist auch Q wahr.

Analog zum Beispiel:
- P sei "es regnet", Q sei "straße ist nass".
- Wenn nun P eintritt, dann folgt logisch Q

**modus tollens**:

Das ganze gilt Negiert: Wenn ¬P ,dann folgt ¬Q.

- Wenn es '**nicht regnet**', dann ist '**die straße nicht nass**'.

**Resolutionsregel**:

Darüber hinaus lassen sich Beziehung Auflösen (resultieren in..):

- Wenn **P → Q** ist und **Q → R** ist, dann gilt **P → R**.

Beispiel:

- Wenn P→Q:
 - **Pflanzenfresser(X) → Säugetier(X)**
- Und Q→R:
 - **Säugetier(X) → Tier(X)**
- Dann gilt P→R:
 - **Pflanzenfresser(X) → Tier(X)**
 
 Oder in die Formel eingesetzt:

$$\frac{Pflanzenfresser(X) → Säugetier(X), Säugetier(X) → Tier(X)}{Pflanzenfresser(X) → Tier(X)}$$

## Suchstrategien

Wenn wir regeln aufstellen in Form von P→Q,
dann wird stets eine Reihenfolge der logischen Zusammenhänge abgearbeitet.

In der logischen Programmierung gibt es daher 2 Suchstrategien:  
**Tiefensuche** und **Breitensuche**

### Tiefensuche
- leicht zu implementieren
- geringer Speicherbedarf
- Abarbeitung wie in anderen Sprachen
- benötigt Backtracking
- unvollständiges Verfahren

<!--  -->


### Breitensuche

- ohne BT
- vollst. Verf.
- von PS abw. Abarb. - kompliz. Verwalt.
- hoher Speicherbedarf
